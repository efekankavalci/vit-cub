```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
from torchvision.transforms.functional import InterpolationMode
```

```python
# Load an image
image_path = '/content/Belted_Kingfisher_0080_70725.jpg'
image = Image.open(image_path).convert('RGB')
```

```python
transform_1 = transforms.Compose([
    transforms.Lambda(lambda x: x.convert('RGB')),  # Convert grayscale to RGB if needed
    transforms.Resize(384, interpolation=InterpolationMode.BICUBIC),
    transforms.CenterCrop(384),
])

transform_2 = transforms.Compose([
    transforms.Lambda(lambda x: x.convert('RGB')),  # Convert grayscale to RGB if needed
    transforms.Resize((384,384), interpolation=InterpolationMode.BICUBIC),
    transforms.CenterCrop(384),
])
```

```python
transformed_image_1 = transform_1(image)
transformed_image_2 = transform_2(image)
```

```python
def show_images(original, transformed_image_1, transformed_image_2):
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    axes[0].imshow(original)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(transformed_image_1)
    axes[1].set_title('Transformed Image 1')
    axes[1].axis('off')

    axes[2].imshow(transformed_image_2)
    axes[2].set_title('Transformed Image 2')
    axes[2].axis('off')

    plt.show()

# Call the function to display the images
show_images(image, transformed_image_1, transformed_image_2)
```
